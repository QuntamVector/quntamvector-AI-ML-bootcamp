```python
# whatsapp message formatter

print("="*45)
print(" 📲 Whatsapp message formatter..")
print("="*45)


#take input
sender = input("your name: ").strip().title()
reciver = input("Friends name: ").strip().title()
message = input("your message: ").strip()

#Analysis

word_count = len(message.split())
char_count = len(message)
char_no_space = len(message.replace(" ",""))

is_formal = any(word in message.lower() for word in ["sir","ma'am","madam","dear"])


has_emoji = any(char for char in message if ord(char) > 127)

# Format the message
first_word = message.split()[0] if message else ""
preview = message[:30] + "..." if len(message) > 30 else message
print()



print('-'*45)
print(f" From: {sender}")
print(f" To: {reciver}")
print(f" preview: {preview}")
print(f" words: {word_count}")
print(f" Chars: {char_count} (characters without spaces: {char_no_space})")
print(f" Foraml: {'yes' if is_formal else 'No'}")
print(f" Emojis: {'yes' if has_emoji else 'No'}")
print('-'*45)

print(f"\n ☁️ {sender} -> {reciver}")
print(f" {message}")
print(f" ☑️ 10:20")
print()

if word_count <3:
    print(" 💡 TIP: Messgae is very short. Add more deetails..")
if message == message.upper() and len(message) > 5:
    print(" 💡 Tip: Writing in ALL CAPS FEELS LIKE SHOUTING.. 🤣")

if not message.endswith((".","!","?","😒","😊")):
    print(" 💡 Tip: Message has no ending punctuations")



    

```
