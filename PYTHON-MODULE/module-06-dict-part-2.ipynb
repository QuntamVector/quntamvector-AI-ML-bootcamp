```python
menu = {
    "Starter": {"Garlic Bread": 80, "Bruschetta": 60},
    "Main Course": {"Margherita Pizza": 12.99, "Pasta Carbonara": 15.99},
    "Dessert": {"Tiramisu": 8.99, "Panna Cotta": 7.99}
}

for category, items in menu.items():
    print(f"{category}:")
    for item, price in items.items():
        print(f"  {item}: ${price}")
```
