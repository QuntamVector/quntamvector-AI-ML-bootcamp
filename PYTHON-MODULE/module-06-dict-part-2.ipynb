# Dictonory
--> menu:--> it will only consider keys
--> key() --> it will print only keys
--> values() --> it will print onlye values
--> items() --> it will print both keys and values
 


Features:
1. Students stored as dictionary: {roll_no: {name, marks_dict}}
2. Marks stored as dict: {"Maths": 85, "Science": 92, ...}
3. Add student: input roll no, name, 5 subject marks
4. View result: show marks + total + average + grade + pass/fail
5. Class topper: find student with highest average
6. Subject topper: find who scored highest in each subject
7. Set usage: find students who passed ALL subjects (marks >= 35)
8. Comparison: which subjects the class found hardest (lowest average)

Grade logic:
  90+  → A+ | 80-89 → A | 70-79 → B | 60-69 → C | Below 60 → F

```python
students = {}

subjects = ["Maths", "Science", "English", "Social", "Computer"]

# Add Students

n = int(input("How many students? "))

for i in range(n):

    roll_no = input("Enter Roll No: ")
    name = input("Enter Name: ")

    marks = {}

    for subject in subjects:
        marks[subject] = int(input(f"Enter {subject} Marks: "))

    students[roll_no] = {
        "name": name,
        "marks": marks
    }

print("\n===== ALL RESULTS =====")

for roll_no, student in students.items():

    print("\n------------------")
    print("Roll No:", roll_no)
    print("Name:", student["name"])

    total = sum(student["marks"].values())
    average = total / len(subjects)

    for subject, mark in student["marks"].items():
        print(subject, ":", mark)

    if average >= 90:
        grade = "A+"
    elif average >= 80:
        grade = "A"
    elif average >= 70:
        grade = "B"
    elif average >= 60:
        grade = "C"
    else:
        grade = "F"

    passed = True

    for mark in student["marks"].values():
        if mark < 35:
            passed = False
            break

    print("Total:", total)
    print("Average:", round(average, 2))
    print("Grade:", grade)
    print("Result:", "PASS" if passed else "FAIL")


# Class Topper

topper_name = ""
highest_avg = 0

for student in students.values():

    avg = sum(student["marks"].values()) / len(subjects)

    if avg > highest_avg:
        highest_avg = avg
        topper_name = student["name"]

print("\n===== CLASS TOPPER =====")
print(topper_name, "-", round(highest_avg, 2))


# Subject Toppers

print("\n===== SUBJECT TOPPERS =====")

for subject in subjects:

    highest_mark = -1
    topper = ""

    for student in students.values():

        mark = student["marks"][subject]

        if mark > highest_mark:
            highest_mark = mark
            topper = student["name"]

    print(subject, ":", topper, "-", highest_mark)


# Students Passed All Subjects (Set Usage)

passed_students = set()

for student in students.values():

    pass_all = True

    for mark in student["marks"].values():
        if mark < 35:
            pass_all = False
            break

    if pass_all:
        passed_students.add(student["name"])

print("\n===== PASSED ALL SUBJECTS =====")

for student in passed_students:
    print(student)


# Hardest Subject

subject_avg = {}

for subject in subjects:

    total = 0

    for student in students.values():
        total += student["marks"][subject]

    avg = total / len(students)

    subject_avg[subject] = avg

print("\n===== SUBJECT AVERAGES =====")

for subject, avg in subject_avg.items():
    print(subject, ":", round(avg, 2))

hardest = min(subject_avg, key=subject_avg.get)

print("\nHardest Subject:", hardest)
print("Average Marks:", round(subject_avg[hardest], 2))
```
