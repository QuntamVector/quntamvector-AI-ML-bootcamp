```python
## NESTED IF 

## ATM MACHINE SIMULATION

print("="*40)
print("     🏧 WELCOME TO PYTHON BANK ATM")
print("="*40)

accounts = {
    "123456": {"pin": "1234", "balance": 15000, "name": "Pritam"},
    "789615": {"pin": "5678", "balance": 30000, "name": "Manasa"},
    "354647": {"pin": "9012", "balance": 40000, "name": "Rohit"},
}

card_number = input("\nEnter card number: ").strip()

if card_number in accounts:

    account = accounts[card_number]

    # PIN CHECK (MAX 3 ATTEMPTS)
    attempt = 0
    pin_correct = False

    while attempt < 3:

        pin = input("Enter your ATM PIN: ").strip()

        if pin == account["pin"]:
            pin_correct = True
            break

        else:
            attempt += 1
            remaining = 3 - attempt

            if remaining > 0:
                print(f"❌ Wrong PIN. {remaining} attempt(s) remaining")

            else:
                print("🚫 Card blocked due to too many wrong attempts")

    # MENU AFTER SUCCESSFUL LOGIN
    if pin_correct:

        print(f"\n✌🏻 Welcome {account['name']}")

        while True:

            print("\n" + "="*35)
            print("1. Check Balance")
            print("2. Withdraw Cash")
            print("3. Deposit Money")
            print("4. Exit")
            print("="*35)

            choice = input("\nSelect option: ").strip()

            # CHECK BALANCE
            if choice == "1":

                print(f"\n💸 Available Balance: Rs {account['balance']}")

            # WITHDRAW CASH
            elif choice == "2":

                amount_str = input(
                    "Enter amount (multiple of 100): Rs "
                ).strip()

                if not amount_str.isdigit():
                    print("❌ Invalid input. Numbers only")

                else:
                    amount = int(amount_str)

                    if amount <= 0:
                        print("❌ Amount must be positive")

                    elif amount % 100 != 0:
                        print("❌ Amount must be multiple of 100")

                    elif amount > 10000:
                        print("❌ Maximum Rs 10,000 per transaction")

                    elif amount > account["balance"]:
                        print("❌ Insufficient balance")

                    else:
                        account["balance"] -= amount

                        print(f"✅ Rs {amount} withdrawn successfully")
                        print(
                            f"💰 Remaining Balance: Rs {account['balance']}"
                        )

            # DEPOSIT MONEY
            elif choice == "3":

                deposit_str = input(
                    "Enter deposit amount: Rs "
                ).strip()

                if not deposit_str.isdigit():
                    print("❌ Invalid input")

                else:
                    deposit = int(deposit_str)

                    if deposit <= 0:
                        print("❌ Deposit amount must be positive")

                    else:
                        account["balance"] += deposit

                        print(f"✅ Rs {deposit} deposited successfully")
                        print(
                            f"💰 Updated Balance: Rs {account['balance']}"
                        )

            # EXIT
            elif choice == "4":

                print("\n🙏 Thank you for using Python Bank ATM")
                print("👋 Visit Again")
                break

            # INVALID OPTION
            else:
                print("❌ Invalid option selected")

else:
    print("❌ Invalid card number")
```
