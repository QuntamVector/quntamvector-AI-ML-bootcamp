```python
print("🏏 CRICKET BATTING AVERAGE CALCULATOR")
print("="*40)

player_name = input("Enter the player's name: ")
total_runs = int(input("Enter total runs scored: "))
total_innings = int(input("Enter total innings played: "))
not_outs = int(input("Enter number of times not out: "))


#calculating batting average
actual_innings = total_innings - not_outs

batting_average = total_runs / actual_innings

print()
print(f"{player_name}'s Batting Average is: {batting_average:.2f}")
print(f"{total_runs} runs in {actual_innings} innings")
print(f"{total_innings} innings with {not_outs} not outs")


if batting_average >= 50:
    grade = "Excellent 👌"
elif batting_average >= 40:
    grade = "Very Good 👍"
elif batting_average >= 30:
    grade = "Good 🙂"
elif batting_average >= 20:
    grade = "Average 😐"
else:
    grade = "Needs Improvement 😟"

print(f"Performance Grade: {grade}")
```
