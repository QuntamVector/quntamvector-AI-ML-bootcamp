```
Features:
1. Start with an empty watchlist = []
2. Show a menu:
   a) Add movie
   b) Remove movie
   c) Mark as watched (move from watchlist to watched_list)
   d) Show all (watchlist + watched count)
   e) Show stats (total, watched, remaining)
   f) Sort watchlist alphabetically
   g) Exit

3. Use separate lists: watchlist = [] and watched = []
4. When user types a movie name to remove, handle if it's not found
5. Show a recommendation: "You've watched X movies! Great job!"

Sample output:
  🎬 WATCHLIST MANAGER
  ─────────────────────
  Pending  (3): Inception, Interstellar, The Dark Knight
  Watched  (2): Parasite, 3 Idiots
  
  Stats: 5 total | 2 watched | 3 remaining
```


````python
# WATCHLIST MANAGER

watchlist = []
watched = []

while True:
    print("\n🎬 WATCHLIST MANAGER")
    print("-" * 35)
    print("a) Add movie")
    print("b) Remove movie")
    print("c) Mark as watched")
    print("d) Show all")
    print("e) Show stats")
    print("f) Sort watchlist alphabetically")
    print("g) Exit")

    choice = input("\nEnter your choice: ").lower()

    # Add movie
    if choice == "a":
        movie = input("Enter movie name: ").strip()

        if movie in watchlist or movie in watched:
            print(f"'{movie}' already exists.")
        else:
            watchlist.append(movie)
            print(f"'{movie}' added to watchlist.")

    # Remove movie
    elif choice == "b":
        movie = input("Enter movie name to remove: ").strip()

        if movie in watchlist:
            watchlist.remove(movie)
            print(f"'{movie}' removed from watchlist.")
        elif movie in watched:
            watched.remove(movie)
            print(f"'{movie}' removed from watched list.")
        else:
            print(f"'{movie}' not found.")

    # Mark as watched
    elif choice == "c":
        movie = input("Enter movie name watched: ").strip()

        if movie in watchlist:
            watchlist.remove(movie)
            watched.append(movie)
            print(f"'{movie}' marked as watched.")
        else:
            print(f"'{movie}' not found in watchlist.")

    # Show all movies
    elif choice == "d":
        print("\nPending:", watchlist if watchlist else "No pending movies")
        print("Watched:", watched if watched else "No watched movies")

    # Show stats
    elif choice == "e":
        total = len(watchlist) + len(watched)
        watched_count = len(watched)
        remaining = len(watchlist)

        print("\n📊 Stats")
        print(f"Total movies   : {total}")
        print(f"Watched movies : {watched_count}")
        print(f"Remaining      : {remaining}")

        # Recommendation message
        print(f"\n🎉 You've watched {watched_count} movies! Great job!")

    # Sort watchlist
    elif choice == "f":
        watchlist.sort()
        print("Watchlist sorted alphabetically.")

    # Exit
    elif choice == "g":
        print("Exiting Watchlist Manager. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
````
