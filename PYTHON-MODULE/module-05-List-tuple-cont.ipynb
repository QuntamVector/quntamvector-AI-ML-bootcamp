```python
print("="*45)
print("           🍕 Swiggy order simulator")
print("="*45)

MENU = [
    ("Butter chicken", 280),
    ("Panner tikka", 220),
    ("Garlic Nan", 40),
    ("Veg Biryani", 160),
    ("Hot Cake",100),
]


# show the menu
print("\n 🗓️ MENU:")
for i, (item, price) in enumerate(MENU,start=1):
    print(f"{i}.{item} Rs {price}")

#Taking orders:
cart = []
print("➕ ADD ITEMS (enter your number or 0 finsh") 

while True:
    choice=input("Item number: ")

    if not choice.isdigit():
        print("please enter valid number..")
        continue
  
    choice = int(choice)

    if choice == 0:
        break

    if 1 <= choice <= len(MENU):
        selected = MENU[choice -1]
        cart.append(selected)

        print(f" ✅ added: {selected[0]} - Rs{selected[1]}")

    else:
        print(f" ❌ invalid choice. enter 1 to {len(MENU)}")    

if not cart:
    print(f"\n ❌ No items added in the card .. Good Bye")
else:
    subtotal = sum( price for item,price in cart)
    gst = int(subtotal* 0.05)
    delivery = 40
    grand_total = subtotal + gst + delivery
    
    print("\n" + "=" * 45)
    print("      🎁 Final order summary")
    print("\n" + "=" * 45)


    for i, (item,price) in enumerate(cart, start=1):
        print( f"{i}.{item:} Rs{price}")

    
    print("-" * 45)
    print(f"subtotal: rs{subtotal}")
    print(f"gst(5%): Rs{gst}")
    print(f"delivery: Rs{delivery}")
    print(f"grand_total: Rs{grand_total}")
```
