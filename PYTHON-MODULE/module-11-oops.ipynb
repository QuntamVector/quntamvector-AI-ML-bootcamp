```python
class BankAccount:
    # constrctor
    def __init__(self,account_number,account_holder,balance=20000):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    
    def deposit(self,amount):
        self.balance += amount
        print(f"Deposited the {amount} to the account and your new balance is {self.balance}")

    def withdraw(self,amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"withdrwed {amount} from the bank account. new balace is {self.balance}")
        else:
            print("Inssuficient funds...!")
            
    def display_balance(self):
        print(f"Accountnumber:  {self.account_number},AccountHolder{self.account_holder}, Balance is {self.balance} ")

        
```

```python
class Bike():
    def __init__(self,wheels,engine_cap,engine_type):
        self.wheels = wheels
        self.engine_cap = engine_cap
        self.engine_type = engine_type

class Ola(Bike):
    def __init__(self, wheels, engine_cap, engine_type, bike_type):
        super().__init__(wheels, engine_cap, engine_type)
        self.bike_type = bike_type

    def wheels(self):
        print(f"no of wheels {self.wheel}")
```

```python
bike=Ola(2,"150cc","electrical","sports")
print(bike.wheels)
```

```python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("sub class shuld implement this")   

class Pet:
    def __init__(self,owner):
        self.owner = owner

class Dog(Animal,Pet):
    def __init__(self, owner,name):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)

    def speak(self):
        print(f"{self.name} says woff")

    def own(self):
        print(f"{self.owner}")


```

```python
dog=Dog("Manoj","Tommy")
dog.own()
```
