```python
with open("examples/manoj.txt","w") as file:
    file.write("This is a new file created using the with statement.")
    file.write("\nThis file is used to demonstrate the with statement for file handling in Python.")
    file.write("\nThe with statement ensures that the file is properly closed after its suite finishes, even if an exception is raised.")

print("file written")
```

```python
with open("examples/manoj.txt","w") as file:
    file.write("This is a new file created using the with statement.")

print("file written")
```

```python
with open("examples/manoj.txt","a") as file:
    file.write("\nThis is a new file created using the with statement.")
    file.write("\nThis file is used to demonstrate the with statement for file handling in Python.")
    file.write("\nThe with statement ensures that the file is properly closed after its suite finishes, even if an exception is raised.")

print("file written")
```

```python
with open("examples/manoj.txt","r") as file:
    for line_num, line in enumerate(file, start=1):
        print(f"Line {line_num}: {line.strip()}")
```
