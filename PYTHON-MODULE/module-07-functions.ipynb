```python
def calculate_pizza_price(size):
    """
    Calculates the price of a pizza based on its size
    """
  
    if size == "small":
        return 199
    elif size == "medium":
        return 299
    elif size == "large":
        return 399
    else:
        return "Invalid size"

```

```python
calculate_pizza_price('small')
```

```python
price1=calculate_pizza_price('small')
price2=calculate_pizza_price('medium')

print(f"total rupees {price1+price2}")
```
