```python
def calculate_pizza_price(size):
    """
    Calculates the price of a pizza based on its size
    """
  
    if size == "small":
        return 199
    elif size == "medium":
        return 299
    elif size == "large":
        return 399
    else:
        return "Invalid size"

```

```python
# return multiple values
def calculate_bill(subtotal):
    gst = subtotal * 0.18
    delivry_charges = 40
    total = delivry_charges + gst + subtotal

    return subtotal,gst,delivry_charges,total
```

```python
subtotal,gst,delivry_charges,total = calculate_bill(400)
```

```python
print(subtotal)
print(gst)
print(delivry_charges)
print(total)
```

```python
def create_pizza_order(customer_name,size,topping,quantity=1,extra_cheese=False):
    """
    customer_name : requires paramter
    size.         : requires paramter
    topping       : requires paramter
    quantity=1.   : optional default value as 1
    extra_cheese. : optional default is false

    """

    prices = {
        "small" : 199,
        "medium" : 299,
        "large"  : 399,
    }


    base_price = prices.get(size,0)
    topping_cost = 50
    cheese_cost = 30 if extra_cheese else 0

    item_price = base_price + topping_cost +  cheese_cost

    total_price = item_price * quantity

    print(f"🍕 ORDER DETAILS: ")
    print(f"Customer: {customer_name}")
    print(f"Size: {size}")
    print(f"Topping : {topping}")
    print(f"Quantity: {quantity}")
    print(f"Extra-Cheese: {'YES +Rs30' if extra_cheese else 'No'}")
    print(f"Unit Price: Rs{item_price}")
    print(f"Total Price: {total_price}")

```

```python
create_pizza_order("manasa","large","Pepporoni")
```

```python
create_pizza_order("manasa","EXtra-large","Pepporoni")
```

```python
create_pizza_order("manasa","EXtra-large","PEPONIN")
```

```python
create_pizza_order("manasa","large","PEPONIN",2,"YES")
```

```python
create_pizza_order("manasa","large")
```
